```python
import yfinance as yf
import matplotlib.pyplot as plt
import datetime

def plot_ytd_returns(tickers):
    """
    Plots the year-to-date (YTD) share price returns normalized at 100 for January 1, 2024.

    :param tickers: List of ticker symbols to plot.
    """
    start_date = "2024-01-01"
    end_date = datetime.datetime.now().strftime("%Y-%m-%d")

    # Fetch data for all tickers
    price_data = {}
    for ticker in tickers:
        try:
            stock_data = yf.download(ticker, start=start_date, end=end_date)['Adj Close']
            if not stock_data.empty:
                price_data[ticker] = stock_data
        except Exception as e:
            print(f"Could not fetch data for {ticker}: {e}")

    # Normalize data to 100 at the start date
    normalized_data = {}
    for ticker, data in price_data.items():
        normalized_data[ticker] = (data / data.iloc[0]) * 100

    # Plot data
    plt.figure(figsize=(14, 8))
    for ticker, data in normalized_data.items():
        plt.plot(data.index, data, label=ticker)

    plt.title("YTD Share Price Returns Normalized at 100 (January 1, 2024)", fontsize=14)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Normalized Price", fontsize=12)
    plt.legend(loc='upper left', fontsize=10)
    plt.grid(True)
    plt.show()

# Example usage
tickers = ["ENR.DE", "GE", "ABBN", "SU.PA", "HON", "EMR", "ETN", "ROK", "PH", "AME", "IR"]
plot_ytd_returns(tickers)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import datetime

def plot_ytd_returns_to_pdf(tickers, output_file="ytd_returns.pdf"):
    """
    Plots the year-to-date (YTD) share price returns normalized at 100 for January 1, 2024,
    and saves the plot to a PDF file.

    :param tickers: List of ticker symbols to plot.
    :param output_file: Name of the output PDF file.
    """
    start_date = "2024-01-01"
    end_date = datetime.datetime.now().strftime("%Y-%m-%d")

    # Fetch data for all tickers
    price_data = {}
    for ticker in tickers:
        try:
            stock_data = yf.download(ticker, start=start_date, end=end_date)['Adj Close']
            if not stock_data.empty:
                price_data[ticker] = stock_data
        except Exception as e:
            print(f"Could not fetch data for {ticker}: {e}")

    # Normalize data to 100 at the start date
    normalized_data = {}
    for ticker, data in price_data.items():
        normalized_data[ticker] = (data / data.iloc[0]) * 100

    # Plot data
    plt.figure(figsize=(14, 8))
    for ticker, data in normalized_data.items():
        plt.plot(data.index, data, label=ticker)

    plt.title("YTD Share Price Returns Normalized at 100 (January 1, 2024)", fontsize=14)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Normalized Price", fontsize=12)
    plt.legend(loc='upper left', fontsize=10)
    plt.grid(True)

    # Save to PDF
    plt.savefig(output_file, format="pdf")
    plt.close()
    print(f"Plot saved to {output_file}")

# Example usage
tickers = ["ENR.DE", "GE", "ABBN", "SU.PA", "HON", "EMR", "ETN", "ROK", "PH", "AME", "IR"]
plot_ytd_returns_to_pdf(tickers)
```
